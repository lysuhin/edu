### Функция для тестирования сортировок путём сравнения результатов с built-in сортировкой в Python и замеров времени для массивов разного размера и исходного состояния

```python
def test(sorter, case="random"):
    sizes = [2 ** p for p in xrange(0, 15)]
    times = []
    for size in tqdm.tqdm_notebook(sizes, leave="False"):
        if case == "random":
            a = np.random.randint(0, size, size).tolist()
            a_sorted_ground = sorted(a)
        elif case == "ascending":
            a = range(0, size, 1)
            a_sorted_ground = sorted(a)
        elif case == "descending":
            a = range(0, size, 1)[::-1]
            a_sorted_ground = sorted(a)
        t1 = time()
        a_sorted_manual = sorter(a)
        t2 = time()
        passed = (a_sorted_ground == a_sorted_manual)
        assert passed
#         print "%s array of size %-6d: sorting time %-5.6f" % (case, size, t2-t1)
        times.append(t2-t1)
    return times
```

```python
def draw_cases(times, name):
    x = [2 ** p for p in xrange(0, 15)]

    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1)
    plt.xlabel('# of elements in array')
    plt.ylabel('sorting time')
    plt.title("%s sort cases comparison" % name, fontsize=16)
    plt.plot(x, times["random case"], label="random")
    plt.plot(x, times["ascending case"], label="ascending")
    plt.plot(x, times["descending case"], label="descending")
    plt.legend(loc='best')

    plt.subplot(1, 2, 2)
    plt.xlabel('# of elements in array')
    plt.ylabel('sorting time')
    plt.yscale('log')
    plt.xscale('log')
    plt.title("%s sort cases comparison" % name, fontsize=16)
    plt.plot(x, times["random case"], label="random")
    plt.plot(x, times["ascending case"], label="ascending")
    plt.plot(x, times["descending case"], label="descending")
    plt.legend(loc='best')
```

### Bubble Sort

```python
def bubble_sort(a):
    sorted = False
    while not sorted:
        sorted = True
        for j in xrange(len(a)-1):
            if a[j] > a[j+1]:
                a[j], a[j+1] = a[j+1], a[j]
                sorted = False
    return a
```

```python
times_bubble = {"random case" : test(bubble_sort),
                "ascending case" : test(bubble_sort, case="ascending"),
                "descending case" : test(bubble_sort, case="descending")}
```

```python
draw_cases(times_bubble, "Bubble")
```

### Selection Sort

![selection sort trace](support/selsort "selection sort trace")

```python
def selection_sort(a):
    for j in xrange(len(a)):
        minimum = a[j]
        iminimum = j
        for i in xrange(j, len(a)):
            if a[i] < minimum:
                minimum = a[i]
                iminimum = i
        a[j], a[iminimum] = a[iminimum], a[j]
    return a
```

```python
times_selection = {"random case" : test(selection_sort),
                "ascending case" : test(selection_sort, case="ascending"),
                "descending case" : test(selection_sort, case="descending")}
```

```python
draw_cases(times_selection, "Selection")
```

### Insertion Sort

![insertion sort trace](support/inssort "insertion sort trace")

```python
def insertion_sort(a):
    for j in xrange(len(a)):
        while (j > 0 and a[j] < a[j-1]):
            a[j], a[j-1] = a[j-1], a[j]
            j -= 1
    return a
```

```python
times_insertion = {"random case" : test(insertion_sort),
                "ascending case" : test(insertion_sort, case="ascending"),
                "descending case" : test(insertion_sort, case="descending")}
```

```python
draw_cases(times_insertion, "Insertion")
```

### Shell Sort

![shell 3-sort trace](support/shsort "shell 3-sort trace")

```python
def shell_sort(a):
    h = 1
    hs = [1]
    
    while (3 * h + 1 < len(a)):
        h = 3 * h + 1
        hs.append(h)
        
    for h in hs[::-1]:
        for j in xrange(len(a)):
            while (j-h >= 0 and a[j] < a[j-h]):
                a[j], a[j-h] = a[j-h], a[j]
                j -= h
    return a
```

```python
times_shell = {"random case" : test(shell_sort),
                "ascending case" : test(shell_sort, case="ascending"),
                "descending case" : test(shell_sort, case="descending")}
```

```python
draw_cases(times_shell, "Shell")
```

### Merge Sort

![merge sort trace](support/mersort "merge sort trace")

```python
def merge(a, aux, lo, mid, hi):
#     assert (sorted(a[lo:mid]) == a[lo:mid])
#     assert (sorted(a[mid+1:hi]) == a[mid+1:hi])
#     assert (len(a) == len(aux)) 
        
    for j in xrange(lo, hi+1):
        aux[j] = a[j]
    i, j = lo, mid + 1
        
    for k in xrange(lo, hi+1):
        if i > mid:
            a[k] = aux[j]
            j += 1
        elif j > hi:
            a[k] = aux[i]
            i += 1
        elif aux[i] < aux[j]:
            a[k] = aux[i]
            i += 1
        else:
            a[k] = aux[j]
            j += 1
            
#     assert (sorted(aux[lo:hi+1]) == a[lo:hi+1])
    return a

def msort(a, aux=None, lo=None, hi=None):
    
    if hi <= lo:
        return a
        
    mid = lo + (hi - lo) / 2
    a = msort(a, aux, lo, mid)
    a = msort(a, aux, mid + 1, hi)
    a = merge(a, aux, lo, mid, hi)
    
    return a

def merge_sort(a):
    aux = a[:]
    msort(a, aux, 0, len(a) - 1)
    return a
```

```python
times_merge = {"random case" : test(merge_sort),
                "ascending case" : test(merge_sort, case="ascending"),
                "descending case" : test(merge_sort, case="descending")}
```

```python
draw_cases(times_merge, "Merge")
```

### Quick Sort

![quick sort trace](support/qsort "quick sort trace")

```python
def partition(a, lo, hi):
    i, j = lo, hi
    mid = lo
    pivot = a[mid]
    while (1):
        while (a[i] <= pivot):
            i += 1
            if i == hi:
                break
        while (a[j] >= pivot):
            j -= 1
            if j == lo:
                break
        if i >= j:
            break
            
        a[i], a[j] = a[j], a[i]   
    
    a[mid], a[j] = a[j], a[mid]
    return j

def qsort(a, lo, hi):
    if hi <= lo:
        return
    j = partition(a, lo, hi)
    qsort(a, lo, j-1)
    qsort(a, j+1, hi)
    return

def quick_sort(a):
    qsort(a, 0, len(a) - 1)
    return a
```

```python
draw_cases(times_quick, "Quick")
```

### Heap Sort

![heap sort trace](support/heapsort "heap sort trace")

```python
def heap_sort(a, comparator=None):
    
    def __less(a, i, j):
        if comparator is None:
            return a[i - 1] < a[j - 1]
        else:
            return comparator(a[i - 1], a[j - 1])
    
    def __exch(a, i, j):
        a[i - 1], a[j - 1] = a[j - 1], a[i - 1]
    
    def __swim(a, ind, N):
        while (ind > 1 and __less(a, ind / 2, ind)):
            __exch(a, ind, ind / 2)
            ind = ind / 2
               
    def __sink(a, ind, N):
        while (2 * ind <= N):
            j = 2 * ind
            if (j < N and __less(a, j, j+1)):
                j += 1
            if not __less(a, ind, j):
                break
            __exch(a, ind, j)
            ind = j
    
    def __heapify(a):
        N = len(a)
        ks = xrange(N / 2, 0, -1)
        for k in ks:
            __sink(a, k, N)
        return a
    
    def __sort(a):
        a = __heapify(a)
        N = len(a)
        while N > 1:
            __exch(a, 1, N)
            N -= 1
            __sink(a, 1, N)
        return a
    
    return __sort(a)
```

```python
times_heap = {"random case" : test(heap_sort),
                "ascending case" : test(heap_sort, case="ascending"),
                "descending case" : test(heap_sort, case="descending")}
```

```python
draw_cases(times_heap, "Heap")
```
