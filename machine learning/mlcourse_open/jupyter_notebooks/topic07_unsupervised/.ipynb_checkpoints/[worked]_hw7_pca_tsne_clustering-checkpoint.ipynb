Для PCA в данной задаче стоит использовать параметр [`svd_solver='randomized'`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html), который является приближением PCA, но позволяет существенно ускорить его работу на больших наборах данных. Для того, чтобы результаты можно было сравнивать, нужно зафиксировать `random_state=1`.

```python
X_sc = StandardScaler().fit_transform(lfw_people.data)
```

```python
pca = PCA(n_components=100, random_state=1, svd_solver='randomized')
explained_vars = pca.fit(X_sc).explained_variance_ratio_
```

```python
explained_vars_cumsum = np.cumsum(explained_vars)
explained_vars_cumsum
```

```python
plt.step(range(100), explained_vars_cumsum, label='explained variance')
plt.xlabel('$N_{components}$')
plt.hlines(0.9, 0, 100)
```

```python
np.where(explained_vars_cumsum > 0.9)
```

```python
components = pca.components_[:30].reshape((30, 50, 37))
```

```python
fig = plt.figure(figsize=(15, 6))

for i in range(30):
    ax = fig.add_subplot(3, 10, i + 1, xticks=[], yticks=[])
    ax.imshow(components[i], cmap='gray')
plt.show()
```

```python
pca = PCA(n_components=2, svd_solver='randomized', random_state=1)
pca.fit(X_sc)
```

```python
projections = {}
for j, name in enumerate(lfw_people.target):
    x1x2 = pca.transform(X_sc[j])[0]
    x1 = x1x2[0]
    x2 = x1x2[1]
#     x1 = np.dot(X_sc[j], pca.components_[0])
#     x2 = np.dot(X_sc[j], pca.components_[1])
    if name in projections.keys():
        projections[name].append([x1, x2])
    else:
        projections.update({name : [[x1, x2]]})
```

```python
mean_projections = {name : [np.array(projections_list).mean(axis=0)[0], np.array(projections_list).mean(axis=0)[1]] for name, projections_list in projections.items()}
```

```python
lfw_people.target_names
```

```python
for name, (x1, x2) in mean_projections.items():
    plt.scatter(x1, x2)
    plt.annotate(lfw_people.target_names[name], (x1, x2))
```

```python
from scipy.spatial import distance_matrix
```

```python
dmat = distance_matrix(np.array(mean_projections.values()), np.array(mean_projections.values()))
```

В данном случае мы ищем наиболее выделеный излом в кривой расстояний до цетроидов в зависимости от числа кластеров. Будем рассматривать возможное число кластеров от 2 до 10. Как и в предыдущей задаче, нужно будет использовать `random_state=1` для инициализации алгоритма k-means.

```python
ks = range(1, 10, 1)
inertias = []
for k in ks:
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(ks, inertias)
```

Вернемся к набору данных с лицами. Притворимся, что мы не знаем ответов – на какой фотографии кто изображен, но знаем, что это 12 разных людей. Давайте сравним результаты кластеризации фотографий с помощью 4 алгоритмов – KMeans, агломеративной кластеризации, AffinityPropagation и спектральной кластеризации. Параметры алгоритмов возьмите как в примере в конце [статьи](https://habrahabr.ru/company/ods/blog/325654/), только измените число кластеров на 12.

```python
algorithms = []
algorithms.append(KMeans(n_clusters=12, random_state=1))
algorithms.append(AffinityPropagation())
algorithms.append(SpectralClustering(n_clusters=12, random_state=1,
                                     affinity='nearest_neighbors'))
algorithms.append(AgglomerativeClustering(n_clusters=12))
```

```python
data =[]
y = lfw_people.target
for algo in algorithms:
    algo.fit(X_sc)
    data.append(({
        'ARI': metrics.adjusted_rand_score(y, algo.labels_),
        'AMI': metrics.adjusted_mutual_info_score(y, algo.labels_),
        'Homogenity': metrics.homogeneity_score(y, algo.labels_),
        'Completeness': metrics.completeness_score(y, algo.labels_),
        'V-measure': metrics.v_measure_score(y, algo.labels_),
        'Silhouette': metrics.silhouette_score(X_sc, algo.labels_)}))

results = pd.DataFrame(data=data, columns=['ARI', 'AMI', 'Homogenity',
                                           'Completeness', 'V-measure', 
                                           'Silhouette'],
                       index=['K-means', 'Affinity', 
                              'Spectral', 'Agglomerative'])

results
```

```python
algorithms_2 = []
algorithms_2.append(KMeans(n_clusters=2, random_state=1))
algorithms_2.append(AffinityPropagation())
algorithms_2.append(SpectralClustering(n_clusters=2, random_state=1,
                                     affinity='nearest_neighbors'))
algorithms_2.append(AgglomerativeClustering(n_clusters=2))
```

```python
data =[]
y = (lfw_people.target == 10).astype(int)
for algo in algorithms_2:
    algo.fit(X_sc)
    data.append(({
        'ARI': metrics.adjusted_rand_score(y, algo.labels_),
        'AMI': metrics.adjusted_mutual_info_score(y, algo.labels_),
        'Homogenity': metrics.homogeneity_score(y, algo.labels_),
        'Completeness': metrics.completeness_score(y, algo.labels_),
        'V-measure': metrics.v_measure_score(y, algo.labels_),
        'Silhouette': metrics.silhouette_score(X_sc, algo.labels_)}))

results = pd.DataFrame(data=data, columns=['ARI', 'AMI', 'Homogenity',
                                           'Completeness', 'V-measure', 
                                           'Silhouette'],
                       index=['K-means', 'Affinity', 
                              'Spectral', 'Agglomerative'])

results
```
