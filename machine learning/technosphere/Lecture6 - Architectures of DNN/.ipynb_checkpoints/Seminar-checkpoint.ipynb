 Задание: выкинуть последний слой из vgg / alexnet и обучить линейную регрессиию на выходах полученных сетей.
Например, для mnist.

Отрежем от сети только ту часть, которая отвечает за свёртки:

```python
#transfer_features = torchvision.models.vgg.vgg16(pretrained=True).features
transfer_features = torchvision.models.alexnet(pretrained=True).features
```

Запретим весам обновляться:

```python
for layer in transfer_features:
    for parameter in layer.parameters():
        parameter.requires_grad = False
```

```python
import numpy as np
from sklearn.datasets import fetch_mldata
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from torch import nn
from torch.autograd import Variable
%matplotlib inline
```

Загрузим данные:

```python
# обратите внимание, что данные имеют размерость examples x channels x w x h
mnist = fetch_mldata('MNIST original')
images = mnist['data']
images = images.reshape(images.shape[0], 1, 28, 28)
labels = mnist['target'].astype(int)
np.random.seed(777)
images = np.stack((images, images, images), axis=1).reshape((70000, 3, 28, 28))
images_train, images_test, labels_train, labels_test = train_test_split(images, labels)
for i in xrange(25):
    plt.subplot(5, 5, i+1)
    plt.imshow(images_train[i, 0, :, :], cmap=plt.cm.Greys_r)
    plt.axis('off')
del images, labels
```

Напишем функцию для перевода картинки в нужный размер:

```python
def to_224(image):
    image_224 = torch.nn.UpsamplingNearest2d(size=(224,224))(image)
    return image_224
```

Напишем слой для сети, который будет автоматически делать upsampling входных картинок:

```python
class Converter(nn.Module):
    def __init__(self):
        super(Converter, self).__init__()
        
    def forward(self, x):
        output = to_224(x)
        return output
```

Напишем архитектуру нашей сети: рукописный umpsampling + признаки из AlexNet + линейная регрессия + LogSoftMax():

```python
class MyTransferLearning(nn.Module):
    def __init__(self):
        super(MyTransferLearning, self).__init__()
        self.resize_layer = Converter()
        self.feature_ext_layers = transfer_features
        self.linear_layers = nn.Sequential(#nn.Linear(512 * 7 * 7, 10), 
                                            nn.Linear(256 * 6 * 6, 10),
                                            nn.LogSoftmax())
        
    def forward(self, x):
        x = self.resize_layer(x)
        x = self.feature_ext_layers(x)
        x = x.view(x.size(0), -1)
        x = self.linear_layers(x)
        return x
    
    def predict_proba(self, x):
        return torch.exp(self.forward(x))
    
    def predict(self, x):
        return self.predict_proba(x).data.numpy().argmax(axis=1)
```

```python
network = MyTransferLearning()
```

Напишем функцию для обучения сети:

```python
def train_network(network, X, Y, loss_type='NLL', n_epochs=10, batch_size=10, opt_type='SGD', lr=0.001, verbose=1, 
                 l2_coeff=0.):
    
    if loss_type == 'NLL': 
        loss_fn = torch.nn.NLLLoss()
    if opt_type == 'SGD': 
        optimizer = torch.optim.SGD(network.linear_layers.parameters(), lr=lr, weight_decay=l2_coeff)
    
    loss_avg = 0
    losses_epochs = []
    
    from tqdm import tqdm
    from sklearn.utils import shuffle
    
    n_batches = X.shape[0] / batch_size
    dtype = torch.FloatTensor
    
    for epoch in tqdm(xrange(n_epochs)):
        
        if epoch % 5 == 0:
            state_dict = optimizer.state_dict()
            state_dict['param_groups'][0]['lr'] /= 2.
            optimizer.load_state_dict(state_dict)
        
        X_train, Y_train = shuffle(X, Y)
        losses = []
        for batch in xrange(n_batches):
            x = Variable(torch.from_numpy(X_train[batch : batch + batch_size]).type(dtype), requires_grad = False)
            y = Variable(torch.from_numpy(Y_train[batch : batch + batch_size]).type(torch.LongTensor), requires_grad = False)
        
            y_pred = network.forward(x)
            loss = loss_fn(input = y_pred, target = y)
            losses.append(loss.data[0])
        
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
        
        losses_epochs.append(np.mean(losses))
        
        if verbose: print ("%d, %.4f" % (epoch, losses_epochs[-1]))#, acc_test))
    return network, losses_epochs
```

Обучимся на 20% данных (долго, очень долго - целый час):

```python
network, losses = train_network(network, images_train[:10000, ...], labels_train[:10000], l2_coeff=1e-4)
```

Изобразим лосс по мере обучения:

```python
plt.plot(losses)
```

Сохраним модель / Загрузим сохраненную:

```python
import pickle as pkl
with open('network_alexnet.pkl', 'rb') as f:
    network = pkl.load(f)
#with open('network_alexnet.pkl', 'wb') as f:
#    pkl.dump(network, f, pkl.HIGHEST_PROTOCOL)
```

Практика показала, что делать предсказание для всего датасета сразу - самоубийство по памяти. Напишем функцию для предсказания по батчам:

```python
def make_batch_prediction(model, x, batch_size=10):
    predictions = []
    from tqdm import tqdm
    for i in tqdm(xrange(x.shape[0] / batch_size)):
        y_pred = model.predict(Variable(torch.from_numpy(images_test[i*batch_size:(i+1)*batch_size]), requires_grad=False).type(dtype))
        predictions.extend(y_pred.tolist())
        del y_pred
    return predictions
```

```python
y_pred = make_batch_prediction(network, images_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(labels_test, y_pred)
```

Получаем точность на тесте выше 95%. Это, конечно, лучше, чем просто логрегрессия, но все равно хуже, чем было при написании своей сети с несколькими полносвязными слоями. Попробуем дообучиться на всем датасете (да, это будет долго, очень-очень долго):

```python
network_full, losses_full = train_network(network, images_train, labels_train, l2_coeff=1e-4)
```

Либо при *transfer learning* все же стоит оставлять возможность оптимизации весов фильтров в сверточных слоях; но это отнимает очень много времени и ресурсов.
