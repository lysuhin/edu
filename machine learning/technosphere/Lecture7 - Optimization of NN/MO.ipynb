```python
def plot_opt(fs, gs, title):
    plt.figure(figsize=(16, 5))
    plt.title(title, fontsize=16)
    plt.plot(fs, label = 'f(x)')
    plt.plot(gs, label = 'grad_f(x)')
    plt.legend()
    plt.show()
```

```python
def GD(func, grad, X, lr=1e-3, eps=1e-5):
    fs = []
    gs = []
    while np.abs(np.linalg.norm(grad(X))) > eps:
        X = X - lr * grad(X)
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

```python
def Momentum(func, grad, X, lr=1e-3, gamma = 0.9, eps=1e-5):
    fs = []
    gs = []
    v = np.ones(grad(X).shape) * 2 * eps;
    while np.abs(np.linalg.norm(grad(X))) > eps:
        v = gamma * v + lr * grad(X)
        X = X - v
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

```python
def NAG(func, grad, X, lr=1e-3, gamma = 0.9, eps=1e-5):
    fs = []
    gs = []
    v = np.ones(grad(X).shape) * 2 * eps
    while np.abs(np.linalg.norm(grad(X))) > eps:
        v = gamma * v + lr * grad(X - gamma * v)
        X = X - v
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

```python
def Adagrad(func, grad, X, lr=0.1, gamma = 0.9, eps=1e-5, eps_smooth=1e-4):
    fs = []
    gs = []
    grad_squred_sum = eps_smooth * np.ones(shape=grad(X).shape)
    dX = np.ones(grad(X).shape) * 2 * eps
    while np.abs(np.linalg.norm(grad(X))) > eps:
        grad_squred_sum += np.square(grad(X))
        dX = lr * (np.divide(grad(X), np.sqrt(grad_squred_sum)))
        X = X - dX
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

```python
def RMSProp(func, grad, X, lr=0.1, gamma = 0.9, eps=1e-5, eps_smooth=1e-4):
    fs = []
    gs = []
    grad_squred_sum = eps_smooth * np.ones(shape=grad(X).shape)
    dX = np.ones(grad(X).shape) * 2 * eps
    E_t = np.zeros(shape=grad(X).shape)
    while np.abs(np.linalg.norm(grad(X))) > eps:
        E_t = gamma * E_t + (1 - gamma) * np.square(grad(X))
        RMS = np.sqrt(E_t)
        dX = lr * (np.divide(grad(X), RMS))
        X = X - dX
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

```python
def Adadelta(func, grad, X, lr=0.1, gamma = 0.9, eps=1e-5, eps_smooth=1e-4):
    fs = []
    gs = []
    dX = np.ones(grad(X).shape) * 2 * eps
    Eg_t = eps_smooth * np.ones(shape=grad(X).shape)
    Ex_t = eps_smooth * np.ones(shape=grad(X).shape)
    while np.abs(np.linalg.norm(grad(X))) > eps:
        Eg_t = gamma * Eg_t + (1 - gamma) * np.square(grad(X))
        RMSg = np.sqrt(Eg_t)
        Ex_t = gamma * Ex_t + (1 - gamma) * dX
        RMSx = np.sqrt(Ex_t)
        dX = (np.divide(RMSx, RMSg))
        X = X - dX
        fs.append(func(X))
        gs.append(grad(X))
    return fs, gs
```

<i> 1.2 Сравните эти методы на функции $J(x, y) = x^2+y^2$</i>

```python
X_0 = 10 * np.array(np.random.random_sample(2))
```

```python
J = lambda X: X[0] ** 2 + X[1] ** 2
grad_J = lambda X: np.array([2.*X[0], 2.*X[1]])
```

```python
print J(X_0)
print grad_J(X_0)
```

```python
fsGD, gsGD = GD(J, grad_J, X_0)
```

```python
plot_opt(fsGD, gsGD, title="GD")
```

```python
fsM, gsM = Momentum(J, grad_J, X_0)
```

```python
plot_opt(fsM, gsM, "Momentum")
```

```python
fsNAG, gsNAG = NAG(J, grad_J, X_0)
```

```python
plot_opt(fsNAG, gsNAG, "NAG")
```

```python
fsAdagrad, gsAdagrad = Adagrad(J, grad_J, X_0)
```

```python
plot_opt(fsAdagrad, gsAdagrad, "Adagrad")
```

```python
fsRMS, gsRMS = RMSProp(J, grad_J, X_0)
```

```python
plot_opt(fsRMS, gsRMS, "RMSProp")
```

```python
fsAdadelta, gsAdadelta = Adadelta(J, grad_J, X_0)
```

```python
plot_opt(fsAdadelta, gsAdadelta, "Adadelta")
```
