```python
import numpy as np
from __future__ import division
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def f(x):
    return np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```python
def plot(f, x):
    plt.plot(xs, map(f, xs), label='f(x)')
    plt.legend()
```

```python
xs = np.arange(1, 30, .5)
```

```python
plot(f, xs)
```

I. Minimize

```python
from scipy.optimize import minimize
```

```python
x_min = minimize(f, 2., method='BFGS')
```

```python
plot(f, xs)
plt.scatter(x_min['x'][0], x_min['fun'])
```

```python
x_min_30 = minimize(f, 30., method='BFGS')
plot(f, xs)
plt.scatter(x_min_30['x'][0], x_min_30['fun'])
```

```python
x_min_30
```

II. Differential Evolution

```python
from scipy.optimize import differential_evolution
```

```python
x_min_de = differential_evolution(f, [(1, 30)])
x_min_de
```

III. Non-smooth function

```python
def g(x):
    return int(f(x))
```

```python
plot(g, xs)
```

```python
min_x_ns_BFGS = minimize(g, 30, method='BFGS')
min_x_ns_BFGS
```

```python
min_x_ns_de = differential_evolution(g, [(1, 30)])
min_x_ns_de
```
